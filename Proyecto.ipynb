```python
# prompt: Mediante el marco de datos Datos_n: Realice un headmap con las variables Tipo Venta y Marca

import matplotlib.pyplot as plt
import seaborn as sns

# Crear una tabla de contingencia
contingencia = Datos_n.groupby(['Tipo Venta', 'Marca']).size().unstack(fill_value=0)

# Crear el mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(contingencia, annot=True, fmt='d', cmap='PuBuGn')
plt.title('Mapa de Calor: Tipo Venta vs Marca')
plt.xlabel('Marca')
plt.ylabel('Tipo Venta')
plt.show()
```

```python
# 6. ¿Que Tipo Venta trae más ingresos, Porta o VR?
#     (Porta=Portabilidad | VR=VentaRegular)

# Filtrar los datos para los tipos de venta "Porta" y "VR"
datos_filtrados = df[df['Tipo Venta'].isin(['Porta', 'VR'])]

# Agrupar por 'Tipo Venta' y sumar los ingresos
ingresos_por_tipo_venta = datos_filtrados.groupby('Tipo Venta')['Ingreso'].sum().reset_index()

# Ordenar por ingresos de mayor a menor
ingresos_por_tipo_venta = ingresos_por_tipo_venta.sort_values(by='Ingreso', ascending=False)

# Mostrar el resultado
print(ingresos_por_tipo_venta)
```

# `Regresión logística`


### OPCION 1: Hacer un group by de meses e ingresos y proyectar el mes siguiente

#### 1. Convertir las fechas a un formato de datetime si es necesario.


```python
# 1.Convertir las fechas a un formato de datetime si es necesario.
# Convertir la columna 'Periodo' a datetime
df['Periodo'] = pd.to_datetime(df['Periodo'].astype(str) + '01', format='%Y%m%d')

# Verificar la conversión
print(df['Periodo'].head())
```

#### 2. Agrupar los datos por mes y calcular el ingreso total para cada mes.

```python
# Agrupar los datos por mes y calcular el ingreso total para cada mes
ingresos_mensuales = df.groupby(df['Periodo'].dt.to_period('M'))['Ingreso'].sum().reset_index()

# Convertir la columna 'Periodo' de nuevo a datetime
ingresos_mensuales['Periodo'] = ingresos_mensuales['Periodo'].dt.to_timestamp()

# Verificar el DataFrame resultante
print(ingresos_mensuales.head())
```

#### 3. Utilizar un modelo de serie temporal para proyectar el ingreso del mes siguiente.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Dividir los datos en variables independientes (X) y dependientes (y)
X = ingresos_mensuales[['Mes_Numero']]
y = ingresos_mensuales['Ingreso']

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)
```

```python
# Predecir el ingreso para el próximo mes
ultimo_mes = ingresos_mensuales['Mes_Numero'].max() + 1
prediccion = model.predict(pd.DataFrame([[ultimo_mes]], columns=['Mes_Numero']))
print(f'Predicción de ingreso para el próximo mes: {prediccion[0]}')
```

```python
# Evaluar el modelo
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
import matplotlib.pyplot as plt

# Visualizar los ingresos históricos y la predicción
plt.figure(figsize=(10, 6))
plt.plot(ingresos_mensuales['Periodo'], ingresos_mensuales['Ingreso'], label='Ingreso histórico')
plt.axvline(ingresos_mensuales['Periodo'].iloc[-1], color='r', linestyle='--', label='Fecha actual')
plt.scatter(ingresos_mensuales['Periodo'].iloc[-1] + pd.DateOffset(months=1), prediccion[0], color='g', zorder=5, label='Predicción')
plt.title('Ingresos Mensuales y Predicción para el Próximo Mes')
plt.xlabel('Fecha')
plt.ylabel('Ingreso')
plt.legend()
plt.grid(True)
plt.show()
```

>**Observación**
> - Se observa en el gráfico los ingresos mensuales desde enero de 2024 hasta abril de 2024.
> - Se puede observar una disminución en los ingresos desde enero hasta marzo de 2024, seguido de un aumento en abril de 2024.
> - Se puede observar en el punto verde la predicción del ingreso para el próximo mes.
